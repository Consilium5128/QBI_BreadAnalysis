```python
# Filtering by pore volume (<1e6 and >25)
bread_metrics_dict["Ro1a"] = bread_metrics_dict["Ro1a"][(bread_metrics_dict["Ro1a"]["volume_ps"]<1e6) & (bread_metrics_dict["Ro1a"]["volume_ps"]>25)]
bread_metrics_dict["Ru1a"] = bread_metrics_dict["Ru1a"][(bread_metrics_dict["Ru1a"]["volume_ps"]<1e6) & (bread_metrics_dict["Ru1a"]["volume_ps"]>25)]
bread_metrics_dict["V1a"] = bread_metrics_dict["V1a"][(bread_metrics_dict["V1a"]["volume_ps"]<1e6) & (bread_metrics_dict["V1a"]["volume_ps"]>25)]
bread_metrics_dict["W1a"] = bread_metrics_dict["W1a"][(bread_metrics_dict["W1a"]["volume_ps"]<1e6) & (bread_metrics_dict["W1a"]["volume_ps"]>25)]
```

```python
# Add sphericity metric
sphericity_pore = getattr(pores_pore_regionprops, "sphericity_pore")

bread_metrics_dict["Ro1a"]["sphericity"] = bread_metrics_dict["Ro1a"].apply(lambda row: sphericity_pore(row["volume_ps"], row["surface_area"]), axis=1)
bread_metrics_dict["Ru1a"]["sphericity"] = bread_metrics_dict["Ru1a"].apply(lambda row: sphericity_pore(row["volume_ps"], row["surface_area"]), axis=1)
bread_metrics_dict["V1a"]["sphericity"] = bread_metrics_dict["V1a"].apply(lambda row: sphericity_pore(row["volume_ps"], row["surface_area"]), axis=1)
bread_metrics_dict["W1a"]["sphericity"] = bread_metrics_dict["W1a"].apply(lambda row: sphericity_pore(row["volume_ps"], row["surface_area"]), axis=1)
```

```python
kmeans= KMeans(4)
x=kmeans.fit_predict(test_df_pore[['label','centroid-0','centroid-1','centroid-2','volume_ps','sphericity']].values)
test_df_pore['group']=x.tolist()
test_df_pore.sample(5,random_state=30)
```

```python
sum(test_df_pore["group"]!=0)
```

```python
# Normalisation
test_df_pore['nlabel'] = (test_df_pore['label']-test_df_pore['label'].mean())/test_df_pore['label'].std()
test_df_pore['ncentroid-0'] = (test_df_pore['centroid-0']-test_df_pore['centroid-0'].mean())/test_df_pore['centroid-0'].std()
test_df_pore['ncentroid-1'] = (test_df_pore['centroid-1']-test_df_pore['centroid-1'].mean())/test_df_pore['centroid-1'].std()
test_df_pore['ncentroid-2'] = (test_df_pore['centroid-2']-test_df_pore['centroid-2'].mean())/test_df_pore['centroid-2'].std()
test_df_pore['nvolume_ps'] = (test_df_pore['volume_ps']-test_df_pore['volume_ps'].mean())/test_df_pore['volume_ps'].std()
test_df_pore['nsphericity'] = (test_df_pore['sphericity']-test_df_pore['sphericity'].mean())/test_df_pore['sphericity'].std()
```

```python
kmeans= KMeans(4)
x=kmeans.fit_predict(test_df_pore[['nlabel','ncentroid-0','ncentroid-1','ncentroid-2','nvolume_ps','nsphericity']].values)
test_df_pore['group']=x.tolist()
test_df_pore.sample(5,random_state=30)
```

```python
metrics_to_compare = ['volume_ps', 'surface_area', 'sphericity']
sampling_size = 3000
for metric in metrics_to_compare:
    print(f'Comparing {metric} \n')
    for i,bread1 in enumerate(bread_types):
        for j,bread2 in enumerate(bread_types):
            if j>i:
                w_stat, p_value = wilcoxon(bread_metrics_dict[bread1][metric].sample(sampling_size), bread_metrics_dict[bread2][metric].sample(sampling_size))
                print(f"WS-Rank Test between {bread1} and {bread2}: W-statistic = {w_stat:.3f}, p-value = {p_value:.4e}")
                print(f'p-value after Bonferonni correction (*{sampling_size}): {p_value*3000:.3e} (0.05)\n')
```

```python
metrics_to_compare = ['nvolume_ps', 'nsurface_area', 'nsphericity']
sampling_size = 3000
for metric in metrics_to_compare:
    print(f'Comparing {metric} \n')
    for i,bread1 in enumerate(bread_types):
        for j,bread2 in enumerate(bread_types):
            if j>i:
                w_stat, p_value = ks_2samp(bread_metrics_dict[bread1]["volume_ps"].sample(sampling_size), bread_metrics_dict[bread2]["volume_ps"].sample(sampling_size))
                print(f"KS-Rank Test between {bread1} and {bread2}: KS-statistic = {w_stat:.3f}, p-value = {p_value:.4e}")
                print(f'p-value after Bonferonni correction (*{sampling_size}): {p_value*3000:.3e} (0.05)\n')
```
